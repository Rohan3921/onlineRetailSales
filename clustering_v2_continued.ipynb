```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```python
df = pd.read_csv('desc_25_11groups.csv')
display(df)
print(df['Group'].value_counts())
missing_group_rows = df[df["Group"].isna()].index
print(len(missing_group_rows))
```

```python
df2 = pd.read_csv('normalized_sales_summary.csv')
df2.rename(columns = {'description':'Description'}, inplace = True)
display(df2)
```

```python
new_df = pd.merge(df, df2, on='Description', how='inner')
display(new_df)
```

```python
import matplotlib.pyplot as plt

# Identify time-based sales columns
time_columns = new_df.columns[6:-2]  # Exclude non-time columns
# print(time_columns)

# Iterate through each group
for group_id in sorted(new_df["Group"].dropna().unique()):  # Drop NaN to avoid errors
    # Select the first 10 items in the group (based on DataFrame order)
    top_items = new_df[new_df["Group"] == group_id].head(10)

    # Create a new figure for each group
    plt.figure(figsize=(12, 6))

    # Plot each itemâ€™s sales trend over time
    for _, row in top_items.iterrows():
        plt.plot(time_columns, row[time_columns], marker="o", label=row["Description"])
        # print(time_columns, row[time_columns])

    # Formatting the plot
    plt.title(f"Group {group_id} - Sales Trend Over Time")
    plt.xlabel("Time (Months)")
    plt.ylabel("Normalized Sales")
    plt.xticks(rotation=45)  # Rotate x-axis labels for readability
    plt.legend(loc="upper right", bbox_to_anchor=(1.3, 1))  # Adjust legend position
    plt.grid(True)

    # Show plot
    plt.show()
```
